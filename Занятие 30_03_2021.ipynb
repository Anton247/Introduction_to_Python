## ОСНОВНЫЕ ОБЪЕКТЫ PANDAS

### SERIES

**Series** — это структура данных, которую можно рассматривать как одну колонку таблицы, одномерный массив. Из предыдущих занятий мы знакомы с похожим одномерным типом данных — списком. Создать **Series** можно:



1.   с помощью команды **pd.Series()**;
2.   pandas.Series(), если библиотека Pandas была импортирована без использования псевдонима.



Создадим Series, где будут содержаться названия первых четырёх месяцев года, и передадим их в виде списка:

```python
import pandas as pd

data = pd.Series(["Январь", "Февраль", "Март", "Апрель"],
                 index = ['Первый', "Второй", "Третий", "Четвёртый"])
display(data)
```

Чтобы сделать вывод данных красивым, используем функцию **display**, на результат вывода это не повлияет.

### ДОСТУП К ЭЛЕМЕНТАМ SERIES

Доступ осуществляется с использованием .loc или .iloc.

.loc принимает определенные метки из индекса, например:

```python
data.loc[["Первый", "Третий"]]
```

.loc можно опустить и писать напрямую data["Первый"] или data[["Первый", "Третий"]].

.iloc принимает на вход порядковые номера элементов Series, например:

```python
 data.iloc[[0, 2]]
```

Если в .loc передать название индекса, а в .iloc номер индекса, которого в данных нет, возникнет ошибка.

### DATAFRAME

Объект DataFrame лучше всего объяснять на примере обычной таблицы. В таблице есть строки и столбцы. Столбцами в объекте DataFrame выступают объекты Series, а их элементы будут строками.

Создать DataFrame проще всего из словаря:

```python
df = pd.DataFrame({'col1': [1, 2], 'col2': [3, 4]})
display(df)
```

Здесь мы создали DataFrame с колонками 'col1' и 'col2' и индексом [0, 1], который сгенерирован автоматически.

Кроме того, DataFrame можно создать из списка списков с указанием названий колонок (columns) и индекс (index).

```python
df = pd.DataFrame([ [1,2,3], [2,3,4] ],
                  columns = ['foo', 'bar', 'baz'],
                  index = ['foobar', 'foobaz'])
display(df)
```

### ПОЛУЧЕНИЕ ИНФОРМАЦИИ О НАБОРЕ ДАННЫХ: DESCRIBE

Метод .describe() показывает основные статистические характеристики данных по каждому **числовому** признаку, это либо int (целые) либо float (вещественные).

```python
football.describe()
```

Для каждого числового столбца (int64) показано число непустых значений (count), среднее (mean), минимум и максимум (min и max), медиана (50%), стандартное отклонение (std) и квартили (25% и 75%). О последних трех характеристиках мы поговорим позднее.

Чтобы посмотреть статистику по нечисловым признакам, нужно явно указать интересующие нас типы в параметре include метода .describe() :

```python
 football.describe(include = ['object'])
```

Если запустить эту ячейку, то мы увидим



*   count - количество объектов
*   uniq - количество уникальных объектов
*   top - самый часто встречающийся объект
*   freq - как часто встречается top





Напомню object - строка

## ИЗВЛЕЧЕНИЕ ДАННЫХ ПО УСЛОВИЯМ

Часто есть необходимость выбрать данные из набора по определенному условию. Например, хотим определить всех футболистов старше 20 лет:


```python
football[football.Age > 20]
```



Синтаксис этой строчки простой: внутри квадратных скобок передается список из True или False и показывает, какие строки DataFrame нужно оставить, а какие убрать.

Или чей возраст больше среднего возраста в датасете

```python
football[football.Age > football.Age.mean()]
```

Но что если условие нас интересует сложное?

Например нас просят определить футболистов ФК Барселона у которых возраст меньше среднего возраста **всех** футболистов, не только из барселоны

```python
football[(football.Age < football.Age.mean()) & (football.Club == "FC Barcelona")]
```

Можно заметить, что мы в скобках указываем каждое отдельное условие. В данном случае первое условие - отбор по возрасту. Затем идёт амперсанд **&** - это логическое И. То есть чтобы оно выполнилось, должны быть истины условия слева и справа от логического И. А второе условие, что клуб равен Барселоне. Равенство - это ДВА равно 

Давайте узнаем что-то конкретное. Например зарплату футболистов барселоны, чей возраст ниже среднего по футболистам

```python
football[(football.Age < football.Age.mean()) & (football.Club == "FC Barcelona")].Wage.mean()
```

4.   Определи название страны (Nationality), из которой больше всего игроков, чья зарплата (Wage) превышает среднее значение. Результат запиши в переменную n.


Выведи результаты работы кода через пробел: speedH, speedL, age, n.

### РЕШЕНИЕ ЗАДАЧИ

```python
import pandas as pd
football = pd.read_csv('football.csv')
speedH = football[football.Wage > football.Wage.mean()].SprintSpeed.mean()
speedH = round(speedH) #round можно было бы написать и на прошлой строке, взяв всё в скобки
speedL = football[football.Wage < football.Wage.mean()].SprintSpeed.mean()
speedL = round(speedL)
age = football[(football.Composure > football.Composure.max() * 0.9) &
               (football.Reactions > football.Reactions.max() * 0.9)].Age.min()
n = football[football.Wage > football.Wage.mean()].Nationality.describe().top
print(speedH, speedL, age, n)
```
